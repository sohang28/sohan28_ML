```python
df = df[['Survived','Age','Sex']]
df = pd.get_dummies(df, columns=['Sex'])
df.dropna(inplace=True)
df.head()
```

## Seperating data into Input and Output

```python
x = df.drop('Survived',axis=1)
x[:5]
```

```python
y = df.Survived
y[:5]
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

## Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

## Testing

```python
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
y_test[:5].values
```

## Performance measures / Metrics

```python
from  sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
# Accuracy Score
accuracy_score(y_test,y_pred)
```

```python
# Model Score
model.score(x_test,y_test)
```

```python
# f1 Score
f1_score(y_test,y_pred)
```

```python
# Recall Score
recall_score(y_test,y_pred)
```

```python
# Precision Score
precision_score(y_test,y_pred)
```

```python
# Confusion matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# check in y_test
y_test
```

```python
sns.heatmap(cm,annot=True,linewidths=.5,cbar=True,cmap = 'summer', xticklabels=['Dead','Survived'],yticklabels=['Dead','Survived'])
plt.xlabel('Predicted_result')
plt.ylabel('Actual_result')
plt.title('Survived Vs Dead')
plt.figure(figsize=(20,12))
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
