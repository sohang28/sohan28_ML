```python
# input: [sweetness,crunchiness]
X = [[10,9],[1,4],[10,1],[7,10],[3,10],[1,1],[5,10]]
```

```python
y = ['fruit','protein','fruit','vegetable','vegetable','protein','vegetable']
```

```python
import numpy as np
np.unique(y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=4)
```

```python
model.fit(X,y)
```

```python
model.classes_
```

```python
model.predict([[3,11]])
```

```python
model.predict_proba([[3,11]])
```

```python
# 4 NN==> 2 (fruits), 2(protein), 3(Veg.)
3/4 # vegetable
```

```python
1/4 # Fruit
```

```python
model.score(X,y)*100
```

## Find the value of K

```python
from sklearn.metrics import f1_score
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X,y)
print(knn.score(X,y)*100)
```

```python
len(X)
```

```python
score = []
for i in range(1,8):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X,y)
    score.append(knn.score(X,y)*100)
```

```python
score
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.plot(range(1,8),score,'--ko',color='Blue')
plt.xlabel('n Neighbors')
plt.ylabel('f1_score')
plt.show()
```

## Conclusion:-
When considering the n-Neighbours(k) parameters 3 and 4, we obtain the same score of 85.714%. In this situation, it is important to assess the number of classes present. In this particular case, there are three classes: Fruit, Protein, and Vegetable. Consequently, the likelihood of clashes occurring is higher when k = 3. Therefore, to mitigate these clashes, we choose k = 4.
